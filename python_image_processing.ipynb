# Importando imagem e obtendo cor de pixel da imagem

```python
from PIL import Image
import os

INPUT_FOLDER = "input"
OUTPUT_FOLDER = "output"

def in_path(filename):
    return os.path.join(INPUT_FOLDER, filename)

def out_path(filename):
    return os.path.join(OUTPUT_FOLDER, filename)
```

# Construindo uma nova imagem

```python
def triangle(x_size, y_size):
    WHITE = (255, 255, 255)
    new_image = Image.new("RGB", (x_size, y_size), WHITE)
    for x in range(x_size):
        for y in range(y_size):
            if x > y:
                new_image.putpixel((x, y), (32, 32, 32))
            elif x < y:
                new_image.putpixel((x, y), (128, 128, 128))
    return new_image

triangulo =  triangle(240, 240)
display(triangulo)
triangulo.save(out_path("triangulo.jpg"))
```

```python
def bandeira_franca(height):
    width = 3 * height // 2

    BLUE = (0, 85, 164)
    WHITE = (255, 255, 255)
    RED = (239, 65, 53)
    
    new_image = Image.new("RGB", (width, height), WHITE)
    
    offset = width // 3

    for x in range(offset):
        for y in range(height):
            if x < height:
                new_image.putpixel((x, y), BLUE)
                new_image.putpixel((x + 2 * offset, y), RED)
    return new_image
    
franca = bandeira_franca(240)
display(franca)
franca.save(out_path("bandeira_franca.jpg"))
```

```python
def bandeira_japao(height):
    width = 3 * height // 2

    WHITE = (255, 255, 255)
    RED = (173, 35, 51)

    raio = 3 * height // 10
    centro = (width // 2, height // 2)

    new_image = Image.new("RGB", (width, height), WHITE)

    for x in range(centro[0] - raio, centro[0] + raio):
        for y in range(centro[1] - raio, centro[1] + raio):
            if (x - centro[0]) ** 2 + (y - centro[1]) ** 2 <= raio ** 2:
                new_image.putpixel((x, y), RED)
                
    return new_image
    
japao = bandeira_japao(240)
display(japao)
japao.save(out_path("bandeira_japao.jpg"))
```

```python
def bandeira_brasil(height):
    width = 3 * height // 2

    GREEN = (0, 156, 59)
    YELLOW = (255, 223, 0)
    BLUE = (0, 39, 118)
    WHITE = (255, 255, 255)

    raio = 3 * height // 10
    centro = (width // 2, height // 2)

    new_image = Image.new("RGB", (width, height), GREEN)

    for x in range(centro[0] - raio, centro[0] + raio):
        for y in range(centro[1] - raio, centro[1] + raio):
            if (x - centro[0]) ** 2 + (y - centro[1]) ** 2 <= raio ** 2:
                new_image.putpixel((x, y), BLUE)
                
    return new_image
    
brasil = bandeira_brasil(240)
display(brasil)
brasil.save(out_path("bandeira_brasil.jpg"))
```

## Convertendo imagem para escala de cinza

```python
def mean_grayscale(image):
    w, h = image.size
    new_image = Image.new("RGB", (w, h))

    for x in range(w):
        for y in range(h):
            rgb = image.getpixel((x, y))
            mean_rgb = sum(rgb) // 3
            new_image.putpixel((x, y), (mean_rgb, mean_rgb, mean_rgb))
    return new_image
```

```python
def grayscale(image):
    w, h = image.size
    new_image = Image.new("RGB", (w, h))

    for x in range(w):
        for y in range(h):
            rgb = image.getpixel((x, y))
            mean_rgb = int(rgb[0] *  0.3 + rgb[1] * 0.59 + rgb[2] * 0.11)
            new_image.putpixel((x, y), (mean_rgb, mean_rgb, mean_rgb))
    return new_image
```

```python
image1_pb = mean_grayscale(image1)
display(image1_pb)
image1_pb.save(out_path("image1_pb.jpg"))

image2_pb = mean_grayscale(image2)
display(image2_pb)
image2_pb.save(out_path("image2_pb.jpg"))
```

```python
image1_pb_media_ponderada = grayscale(image1)
display(image1_pb_media_ponderada)
image1_pb_media_ponderada.save(out_path("image1_pb_media_ponderada.jpg"))

image2_pb_media_ponderada = grayscale(image2)
display(image2_pb_media_ponderada)
image2_pb_media_ponderada.save(out_path("image2_pb_media_ponderada.jpg"))
```
